# Load data and clean data

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
news_df = pd.read_csv('../data/Combined_News_DJIA.csv').drop(labels='Label', axis=1)
news_df['Date'] = pd.to_datetime(news_df['Date'], errors='coerce')
news_df = news_df.set_index('Date')
```

```python
def byte_to_str(value):
    if isinstance(value, str) and (value.startswith("b'") or value.startswith('b"')):
        value = value[2:-1]
    return value

for column in news_df.columns:
    news_df[column] = news_df[column].apply(byte_to_str)
    
news_df.head(3)
```

```python
news_df.to_csv("../data/cleaned_news.csv")
```

```python
import pandas as pd

news_nyt_df = pd.read_csv("../data/response.csv")
news_nyt_df = news_nyt_df.drop_duplicates("abstract")
news_nyt_df
```

```python
import datetime

news_nyt_df['pub_date'] = pd.to_datetime(news_nyt_df['pub_date']).dt.tz_convert('US/Eastern')
news_nyt_df = news_nyt_df.sort_values('pub_date')
start = news_nyt_df.min(axis=0)['pub_date']
end = news_nyt_df.max(axis=0)['pub_date']
start = datetime.datetime(start.year, start.month, start.day)
end = datetime.datetime(end.year, end.month, end.day)
#start = datetime.datetime(2022,3,8)
#end = datetime.datetime(2016,3,29)
print(start)
print(end)
```

```python
from datetime import timedelta

news_df = pd.DataFrame()
date_range = pd.date_range(start, end, tz="US/Eastern")

for row, date in enumerate(date_range):
    day_start = (date - timedelta(days=1)).replace(hour=13, minute=30)
    day_end = date.replace(hour=13, minute=30)
    df_slice = news_nyt_df[(news_nyt_df["pub_date"] >= day_start) & (news_nyt_df["pub_date"] < day_end)]
    if len(df_slice.index) > 0:
        date_index = datetime.datetime(date.year, date.month, date.day)
        news_df.loc[date_index, list(range(len(df_slice.index)))] = df_slice["abstract"].values.T

news_df = news_df.dropna(axis=0, how="all")
news_df
```

```python
import yfinance as yf

tickers = ['^GSPC']
stock_df = yf.download(tickers, start=start,end = end, interval='1d')
stock_df
```

```python
stock_df.to_csv("../data/stock_data.cscv", index=False)
```

```python
data_df = news_df.join(stock_df[['Close']])
data_df = data_df.dropna()
data_df
```

```python
import numpy as np

Y = np.empty((len(data_df.index),))
Y[0:-1] = data_df['Close'].values[1:]
data_df['Label'] = Y
data_df = data_df[:-1]
data_df
```

```python
data_df["Label_Binary"] = np.where(data_df["Close"] < data_df["Label"], 1, 0)
data_df
```

```python
data_df.to_csv("../data/all_data.csv")
```

```python
num_train_items = round(len(data_df.index) * .7)
num_val_items = (len(data_df.index) - num_train_items) // 2
num_test_items = len(data_df.index) - num_train_items - num_val_items
print(num_train_items)
print(num_val_items)
print(num_test_items)
train_df = data_df.iloc[:num_train_items,:]
train_df
```

```python
val_df = data_df.iloc[num_train_items:(num_train_items+num_val_items),:]
val_df
```

```python
test_df = data_df.iloc[(num_train_items+num_val_items):,:]
test_df
```

```python
train_df.to_csv("../data/train_data.csv")
val_df.to_csv("../data/val_data.csv")
test_df.to_csv("../data/test_data.csv")
```
